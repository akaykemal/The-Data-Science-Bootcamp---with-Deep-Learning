```python
df=pd.read_csv("PastHires.csv")
```

```python
df
```

```python
df["Employed?"]=df["Employed?"].map({"Y":1,"N":0})
```

```python
df
```

```python
df=df.replace({"Y":1,"N":0,"BS":1,"MS":2,"PhD":3})
#dictionary'i bu sekilde replace ile bir satir da hepsini degistirdik
```

```python
df
```

```python
df.corr()
```

```python
y=df.Hired
#y=df["Hired"]
```

```python
x=df.drop("Hired",axis=1)
```

```python
x
```

```python
from sklearn.tree import DecisionTreeClassifier
d=DecisionTreeClassifier()
```

```python
d.fit(x,y)
```

```python
df.columns
```

```python
d.predict([[10,1,4,1,0,1]]) # yeni bir isci ekleyip predict yapalim
#bu kisiyi ise alin, cnk 1 cikti, forms doldurtturun sonra outputa gre ise alip almamanizi biliyorsunuz
```

```python
#PCA ile cok veriyi iki seye indirmek istiyorum, suan 6 sutn var, 6 sutna 2'ye dsrebilirmiyim, PCA bu
```

```python
pca=PCA(n_components=2) # number of components=2 --> 2 sutn
```

```python
x_y=pca.fit_transform(x) #x yeni
```

```python
x_y # iki sutna indi ve array oldu, 2 sutn veri ile calisabiliyorum
```

```python
pca.explained_variance_ratio_ #verinin 91% birinci sutnda, 5% ikinci sutnda 
```
