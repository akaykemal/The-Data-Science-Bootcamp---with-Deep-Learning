# Popularity Based Recommender System
 
en cok popüler olan ürünü tavsiye etme

```python
import pandas as pd
```

```python
ratings=pd.read_csv("ml-100k/u.data",sep="\t",header=None,names=["user_id","movie_id","rating"],usecols=range(3))
# csv: Coma Seperated Values
# ancak burdaki sey Tab ile ayrilmis(\t var burda)
# sep (seperator) yazinca virgül yapiyor?
# simdide birinci satir veri -> header=None: ilk satiri baslik gibi almiyor
# suan data ok, ama columns names bilmiyoruz
# names= ile isim veriyoruz ancak son sutün ne bilmiyoruz, onu almiyoruz usecols=range(3), son sütunu almadik
```

```python
ratings.head()
```

```python
movies=pd.read_csv("ml-100k/u.item",sep="|",header=None,encoding="iso-8859-1",names=["movie_id","title"],usecols=range(2))
```

veri tabanin dizayn da takip ettigimis bir sey var:
1. DRY - dont repeat - veri tekrarini engelle

```python
movies.head()
```

```python
ratings=pd.merge(movies,ratings)
```

```python
ratings.head()
```

```python
# en popüler filmi nasil bulacagiz? en cok ismi gecen
# rating ortalama bakinaca mesela bir kisi 5 vermistir, 1 kisiye bakilmaz en az 3 kaynaktan dogrulamak lazim
# kac tane filim var ona bi bakalim
```

```python
ratings.title.nunique()
```

```python
import numpy as np
```

```python
# gruplara ayiriyoruz filimleri
```

```python
grouped=ratings.groupby("title").agg({"rating":[np.size,np.sum,np.mean]})
```

```python
grouped.head()
#size= kac kisi bu filim hakkinda yorum yapmis
#sum= yorumlarin toplam adedi 21
#mean= ortalama ne kadar rating verilmis
```

```python
populer=grouped.sort_values(("rating","mean"),ascending=False)
```

```python
populer.head()
```

```python
toplam=grouped["rating"]["sum"].sum()
```

```python
toplam # yüzdesine göre toplayacam icin hepsini hesapladik
```

```python
populer["percentage"]=populer["rating","sum"].div(toplam)*100
```

```python
populer.head()
```

```python
populer.sort_values(("percentage"),ascending=False)
```
